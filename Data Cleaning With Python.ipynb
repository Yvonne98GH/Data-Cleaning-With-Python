```python
print(upper_limit)
```

```python
print("Upper Limit : {}".format(upper_limit))
print("Lower Limit : {}".format(lower_limit))
```

```python
fig, ax = plt.subplots(figsize = (15,7))
sns.distplot(df['Bitcoin'], ax = ax)

ax.axvline(lower_limit, color = 'red')
ax.axvline(upper_limit, color = 'red')
```

```python
iqr = df['Bitcoin'].quantile(0.75) - df['Bitcoin'].quantile(0.25)
iqr
```

```python
upper_lim = df['Bitcoin'].mean() + 2 * iqr
lower_lim = df['Bitcoin'].mean() - 2 * iqr
```

```python
fig, ax = plt.subplots(figsize = (15,8))
sns.distplot(df['Bitcoin'], ax = ax)
ax.axvline(upper_lim, color = 'red')
ax.axvline(lower_lim, color = 'red')
```
